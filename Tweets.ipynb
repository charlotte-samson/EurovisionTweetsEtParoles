```python
import textblob
import csv
import pandas as pd
import re
from statistics import mean
import matplotlib.pyplot as plt
```

```python
def remove_url(txt):
    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())
```

```python
#ITA
#RUS
#SUI 
#SWE
#NOR
#MKD
#AZE
#AUS
#ISL

"""
./Projet/Top10-2019/Pays-Bas #NED.csv
./Projet/Top10-2019/Italie #ITA.csv
./Projet/Top10-2019/Russie #RUS.csv
./Projet/Top10-2019/Suisse #SUI.csv
./Projet/Top10-2019/Suède #SWE.csv
./Projet/Top10-2019/Norvège #NOR.csv
./Projet/Top10-2019/Macédoine du nord #MKD.csv
./Projet/Top10-2019/Azerbaijan #AZE.csv
./Projet/Top10-2019/Australie #AUS.csv
./Projet/Top10-2019/Islande #ISL.csv
"""
data = pd.read_csv("./Projet/Top10-2019/Islande #ISL.csv",sep=";")
print(data["text"].head)
```

```python
tweets=data["text"]
```

```python
tweets_no_urls = [remove_url(tweet) for tweet in tweets]
```

```python
# Create textblob objects of the tweets
sentiment_objects = [textblob.TextBlob(tweet) for tweet in tweets_no_urls]

sentlist=[]
for sentiment in sentiment_objects :
    sentlist.append(sentiment.polarity)
```

```python
print(sentlist)
```

```python
sentiment_values = [[tweet.sentiment.polarity, str(tweet)] for tweet in sentiment_objects]

# Create dataframe containing the polarity value and tweet text
sentiment_df = pd.DataFrame(sentiment_values, columns=["polarity", "tweet"])

sentiment_df.head()

#répartition 
fig, ax = plt.subplots(figsize=(8, 6))

# Plot histogram of the polarity values
sentiment_df.hist(bins=[-1, -0.75, -0.5, -0.25, 0.25, 0.5, 0.75, 1],
             ax=ax,
             color="purple")

plt.title("Analyse de sentiment des tweets #AUS")
plt.show()
```
